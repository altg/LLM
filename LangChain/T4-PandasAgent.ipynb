```python
import pandas as pd
from langchain.chat_models import ChatOpenAI
from langchain.agents import create_pandas_dataframe_agent
from langchain.agents.agent_types import AgentType
```

```python
input_csv = "../Docs/idb_ops_data.csv"
```

```python
df = pd.read_csv(input_csv)
```

```python
df2 = df[['ops_data_ID', 'operation_id', 'project_code', 'project_title',
       'profile', 'country', 'fund', 'is_member_country',
       'country_code', 'country_ldmc', 'country_class', 'country_region',
       'sector', 'status', 
       'main_mode_of_finance', 'sub_mode_of_finance', 
       'date_of_approval', 'date_of_signature', 
       'date_of_effective', 'date_of_first_disb',
       'date_of_last_disb', 
       'approval_amt_usd',  'disb_amt_usd'
       ]].sample(100)
```

```python
llm = ChatOpenAI(model_name='gpt-3.5-turbo-0613', temperature=0.2)
```

```python

agent = create_pandas_dataframe_agent(llm, df2, 
                                      verbose=True, 
                                      agent_type=AgentType.OPENAI_FUNCTIONS)
  # Perform Query using the Agent
  
```

```python
input_query = "What percentage of projects each year are in LDMC countries?"

response = agent.run(input_query)
```

```python
print(response)
```

```python
import pandas as pd
import numpy as np

# Convert date_of_approval to datetime
df['date_of_approval'] = pd.to_datetime(df['date_of_approval'])

# Extract year from date_of_approval
df['year'] = df['date_of_approval'].dt.year

# Create is_ldmc column
df['is_ldmc'] = np.where(df['country_ldmc'] == 'LDMC', 1, 0)

# Group by year and calculate sum of is_ldmc
ldmc_projects = df.groupby('year')['is_ldmc'].sum()

# Calculate total number of projects for each year
total_projects = df.groupby('year').size()

# Calculate percentage of projects in LDMC countries
percentage_ldmc = (ldmc_projects / total_projects) * 100

# Print the result
print(percentage_ldmc)
```
